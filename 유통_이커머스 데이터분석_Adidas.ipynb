## anova_lm
- anova_lm 함수는 일반적으로 선형 회귀 모델의 분산 분석 결과를 계산하는 데 사용
- 주로 일반적인 선형 모델 (OLS, Ordinary Least Squares)의 분산 분석에 사용




```python
# ols 모형을 아노바 테이블로 바꾸겠다.
anova_tb = anova_lm(model)
print(anova_tb)
# 귀무가설 : 판매방법에 따라 마진율이 같다.
# PR 유의확률 : 0.05 보다 작으면 귀무가설을 기각할 수 있다.
# -> 귀무가설 기각할 수 있다. 0.0 니까
```

```python
# 유의확률에 따른 결론 도출
if anova_tb.iloc[0,4]<0.05:
  print(f'판매방법에 따른 {y}이(가) 차이가 있습니다.')
else:
  print(f'판매방법에 따른 {y}이(가) 차이가 나지 않습니다.')
```

```python
# 위 내용 종합해서 분석 함수 만들기
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

def anova_analysis(x, y):
    # x : 비교그룹(매장, 온라인, 아울렛) -> Sales Method
    # y : 비교할 변수 -> 영업이익, 단가
    df = adidas[[x, y]]


    # 띄어쓰기 있는 경우 statsmodels의 ols 함수가 인식 못함
    # 띄어쓰기를 _로 바꿈
    x = x.replace(' ', '_')
    y = y.replace(' ', '_')

    print('바뀐 이름 : ', x, ':', y)

    # _ 적용한 바뀐 이름으로 columns 수정
    df.columns = [x, y]

    # ols 모델
    # the "C" indicates categorical data
    model = ols(f'{y} ~ C({x})', df).fit()

    # anova_lm 으로 모델 분석
    anova_tb = anova_lm(model)
    print(anova_tb)

    # 유의확률에 따른 결론 도출
    if anova_tb.iloc[0,4]<0.05:
      print(f'판매방법에 따른 {y}이(가) 차이가 있습니다.')
    else:
      print(f'판매방법에 따른 {y}이(가) 차이가 나지 않습니다.')
```

```python
# 판매방법별 단가과 마진율 검증
# 위에서 만든 함수에 적용하기

print('판매가격의 차이 검증')
anova_analysis('Sales Method', 'Price per Unit')
print('\r\n\r\n')
print('====================================================================')
print('\r\n\r\n')
print('마진율의 차이 검증')
anova_analysis('Sales Method', 'Operating Margin')

# 판매방법에 따른 단가와 마진율에 차이가 있다
# 결국은 그래프에서 봤던 것처럼 차이가 있다는 것을 통계적으로 검증 완료
```
